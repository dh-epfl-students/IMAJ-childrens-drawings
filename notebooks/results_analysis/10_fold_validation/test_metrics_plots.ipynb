```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
test_metrics_pt = pd.read_csv("./../../../data/model_learning/comparison/scores/baselines/test_scores_kfold.csv")
test_metrics_pt = test_metrics_pt.replace({"resnext": "Baseline"})

test_metrics_ft_aug = pd.read_csv("./../../../data/model_learning/comparison/scores/fine_tune_aug/test_scores_kfold.csv")

test_metrics_ft_noaug = pd.read_csv("./../../../data/model_learning/comparison/scores/fine_tune_noaug/test_scores_kfold.csv")

test_metrics_pt_replica = pd.read_csv("./../../../data/model_learning/comparison/scores/baselines_replica/test_scores_kfold.csv")
# test_metrics_clus_replica = pd.read_csv("./../../../data/model_learning/comparison/scores/baselines_replica/test_scores_kfold.csv")

test_metrics_mini_replica_ft_aug = pd.read_csv("./../../../data/model_learning/comparison/scores/fine_tune_minireplica_aug/test_scores_kfold.csv")
test_metrics_clus_replica_ft_aug = pd.read_csv("./../../../data/model_learning/comparison/scores/fine_tune_clusreplica_aug/test_scores_kfold.csv")

test_metrics_mini_replica_ft_noaug = pd.read_csv("./../../../data/model_learning/comparison/scores/fine_tune_minireplica_noaug/test_scores_kfold.csv")
test_metrics_clus_replica_ft_noaug = pd.read_csv("./../../../data/model_learning/comparison/scores/fine_tune_clusreplica_noaug/test_scores_kfold.csv")
```

```python
test_metrics_mini_replica = test_metrics_pt_replica[test_metrics_pt_replica["model"]=="Mini-Replica"]
test_metrics_clus_replica = test_metrics_pt_replica[test_metrics_pt_replica["model"]=="Clus-Replica"]

test_ft_baseline = pd.concat([test_metrics_pt, test_metrics_ft_noaug,
                              test_metrics_ft_aug, test_metrics_mini_replica,
                              test_metrics_mini_replica_ft_noaug, test_metrics_mini_replica_ft_aug,
                             test_metrics_clus_replica, test_metrics_clus_replica_ft_noaug,
                             test_metrics_clus_replica_ft_aug])
```

```python
with plt.style.context("default"):
    with plt.xkcd():
        plts = plt.figure(figsize=(15, 9))
        ax1 = plt.subplot2grid(shape=(2,6), loc=(0,0), colspan=2, fig=plts)
        ax2 = plt.subplot2grid((2,6), (0,2), colspan=2, fig=plts, sharey=ax1, sharex=ax1)
        ax3 = plt.subplot2grid((2,6), (0,4), colspan=2, fig=plts, sharey=ax1, sharex=ax1)
        ax4 = plt.subplot2grid((2,6), (1,1), colspan=2, fig=plts, sharey=ax1, sharex=ax1)
        ax5 = plt.subplot2grid((2,6), (1,3), colspan=2, fig=plts, sharey=ax1, sharex=ax1)

        axs = [ax1, ax2, ax3, ax4, ax5]

        recall_columns = ['R@400', 'R@200', 'R@100', 'R@50', 'R@20']

        for num, name in enumerate(recall_columns):

            ax = sns.barplot(x="model", y=name, data=test_ft_baseline, palette="Set3", ax=axs[num], hue="model", dodge=False)
            for i in range(9):
                axs[num].bar_label(axs[num].containers[i], label_type='center', fmt='%.2f')
            axs[num].set_title(name.replace("R", "Recall"))
            # axs[num].tick_params(labelrotation=15)
            axs[num].set_xticks([])
            # axs[num].set_xticks([])
            axs[num].set_ylabel("Recall")

            axs[num].get_legend().remove()


        plts.suptitle("Recall@k")
        handles, labels = ax.get_legend_handles_labels()
        plts.legend(handles, labels, loc='lower right', bbox_to_anchor=(1.01, 0.15))

        # axs[num].legend(loc='center left', bbox_to_anchor=(1, 0.5))
        
        # plt.xticks(np.linspace(0, 20, 100))

        plts.tight_layout()
```

```python
plts = plt.figure(figsize=(11, 10), dpi=500)
ax = sns.boxplot(x="model", y="MP", data=test_ft_baseline, palette="pastel", hue="model", dodge=False, showmeans=False)

means = test_ft_baseline.groupby(['model'])['MP'].mean()
vertical_offset = test_ft_baseline['MP'].mean() * - 0.01 # offset from median for display

for xtick in ax.get_xticks():
    ax.text(xtick, means[xtick] + vertical_offset, np.round(means[xtick], 2), 
            horizontalalignment='center',size='small',color='black',weight='semibold')

ax.set_title("Distribution of Mean Position")
ax.set_xticks([])
ax.set_ylabel("Mean Position")
ax.set_xlabel("Model")
plt.legend(title="Model")
plts.tight_layout()
```

```python
test_ft_best = test_ft_baseline[test_ft_baseline["model"].isin(["Baseline", "FT Aug", "Mini-Replica", "FT Aug Mini-Replica", "Clus-Replica", "FT Aug Clus-Replica"])]
```

```python
plts = plt.figure(figsize=(11, 10), dpi=500)
ax = sns.boxplot(x="model", y="MMP", data=test_ft_best, palette="pastel", hue="model", dodge=False, showmeans=False)

means = test_ft_best.groupby(['model'])['MMP'].mean()
vertical_offset = test_ft_best['MMP'].mean() * - 0 # offset from median for display

for xtick in ax.get_xticks():
    ax.text(xtick, means[xtick] + vertical_offset, np.round(means[xtick], 2), 
            horizontalalignment='center',size='small',color='black',weight='semibold')

ax.set_title("Distribution of Minimum Mean Position")
ax.set_xticks([])
ax.set_ylabel("Minimum Mean Position")
ax.set_xlabel("Model")
plt.legend(title="Model")
plts.tight_layout()
```
